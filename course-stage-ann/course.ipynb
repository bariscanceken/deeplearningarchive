```python
import numpy as np
from sklearn.model_selection import train_test_split
x_l = np.load('X.npy')
Y_l = np.load('Y.npy')
X = np.concatenate((x_l[204:409], x_l[822:1027]), axis=0)
Y = np.concatenate((np.zeros(205), np.ones(205)), axis=0)
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.15, random_state=42)
x_train = x_train.reshape(x_train.shape[0], -1)
x_test = x_test.reshape(x_test.shape[0], -1)
print(f"X_train: {x_train.shape}, y_train: {y_train.shape}")
```

```python
#Evaluating the ANN with Keras
from scikeras.wrappers import KerasClassifier
from sklearn.model_selection import cross_val_score
from keras.models import Sequential
from keras.layers import Dense

def build_classifier():
    classifier = Sequential()
    classifier.add(Dense(units=8,kernel_initializer='uniform',activation='relu', input_dim = x_train.shape[1]))
    classifier.add(Dense(units=4,kernel_initializer='uniform',activation='relu'))
    classifier.add(Dense(units=1,kernel_initializer='uniform',activation='sigmoid'))
    classifier.compile(optimizer='adam', loss= 'binary_crossentropy', metrics= ['accuracy'])
    return classifier

classifier = KerasClassifier(model=build_classifier, epochs=100, batch_size=10)
print("X_train satır sayısı:", x_train.shape[0])
print("y_train satır sayısı:", y_train.shape[0])
accuracies = cross_val_score(estimator = classifier, X= x_train, y=y_train, cv= 3)
mean = accuracies.mean()
variance = accuracies.std()
print("Accuracy Mean :",str(mean))
```
